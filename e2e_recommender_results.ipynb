# Crawling over the e2e recommender results
Use this notebook to synthesize experiment results for the paper.

```python
import os
import json
import pandas as pd
```

```python
e2e_output_dir = 'e2e_output'
rows = {}
for d in os.listdir(e2e_output_dir):
    row = {}
    if os.path.isdir(os.path.join(e2e_output_dir, d)):
        with open(os.path.join(e2e_output_dir, d, 'log.json')) as f:
            data = json.load(f)
            for item in data:
                if 'mAP' in item:
                    row['mAP'] = item['mAP']
                if 'train_accuracies' in item:
                    row['train_accuracy'] = item['train_accuracies'][-1]
                if 'test_accuracies' in item:
                    row['test_accuracy'] = item['test_accuracies'][-1]
                if 'e2e_config' in item:
                    row['learning_rate'] = item['e2e_config']['train']['learning_rate']
                    row['dropout_prob'] = item['e2e_config']['train']['dropout_prob']
                    row['n_clusters'] = item['e2e_config']['data_provider']['cluster_config']['n_clusters']
            rows[d] = row

experiments = pd.DataFrame.from_dict(rows, orient='index')
experiments
```
